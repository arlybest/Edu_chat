```python
import os

def compter_pdfs(dossier):
    """Compte le nombre de fichiers PDF dans un dossier."""
    if not os.path.exists(dossier):
        raise ValueError(f"Le dossier '{dossier}' n'existe pas.")

    pdf_count = len([f for f in os.listdir(dossier) if f.lower().endswith(".pdf")])

    return pdf_count

dossier_cible = "data"  #
nombre_pdfs = compter_pdfs(dossier_cible)
print(f"📂 Nombre de fichiers PDF dans '{dossier_cible}' : {nombre_pdfs}")
```

```python
import os
from langchain_community.document_loaders import UnstructuredPDFLoader
from tqdm.auto import tqdm

# 📌 Pipeline de traitement avec UnstructuredPDFLoader
def traiter_pdf(chemin_pdf: str) -> list[dict]:
    """
    Traite un fichier PDF en utilisant UnstructuredPDFLoader pour extraire le texte.
    """
    loader = UnstructuredPDFLoader(chemin_pdf)
    documents = loader.load()  # Charge le document et extrait le texte

    pages_et_textes = []
    for i, doc in enumerate(documents):
        texte_complet = doc.page_content  # Texte extrait du PDF
        pages_et_textes.append({
            "numero_page": i + 1,
            "type_contenu": "text",  # UnstructuredPDFLoader traite tout comme du texte
            "nb_caracteres": len(texte_complet),
            "nb_mots": len(texte_complet.split(" ")),
            "nb_phrases": len(texte_complet.split(". ")),  # Estimation du nombre de phrases
            "nb_tokens_estime": len(texte_complet) / 4,  # Estimation du nombre de tokens
            "texte": texte_complet
        })

    return pages_et_textes

# 📌 Traitement des fichiers PDF
dossier_pdf = "data"
if not os.path.exists(dossier_pdf):
    raise FileNotFoundError(f"Le dossier '{dossier_pdf}' n'existe pas.")

fichiers_pdf = [os.path.join(dossier_pdf, f) for f in os.listdir(dossier_pdf) if f.endswith(".pdf")]

contenu_tous_les_pdfs = {}
tout_texte = ""

for fichier_pdf in fichiers_pdf:
    print(f"\n📖 Lecture du fichier : {fichier_pdf}")
    contenu_pdf = traiter_pdf(fichier_pdf)
    contenu_tous_les_pdfs[os.path.basename(fichier_pdf)] = contenu_pdf

    for page in contenu_pdf:
        tout_texte += page["texte"] + " "

# 📌 Affichage des résultats
for nom_pdf, contenu in contenu_tous_les_pdfs.items():
    print(f"\n--- {nom_pdf} ---")
    for page in contenu[:2]:  # Affiche les 2 premières pages
        print(f"📄 Page {page['numero_page']} ({page['type_contenu']}):")
        print(page["texte"][:500])  # Affiche les 500 premiers caractères

print("\n📝 Texte complet extrait (1000 premiers caractères) :")
print(tout_texte[:1000])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.schema import Document  # Importation de la classe Document

# Crée un objet de découpeur de texte
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=10000,  # Taille du chunk (en caractères)
    chunk_overlap=200,  # Chevauchement des chunks (en caractères)
    add_start_index=True,  # Suivre l'index dans le document original
)

# Convertir tout_texte en un format que langchain peut utiliser (en tant que Document)
documents = [Document(page_content=tout_texte)]  # Utilise la classe Document

# Divise le texte en sous-documents
all_splits = text_splitter.split_documents(documents)

# Affiche le nombre de sous-documents créés
print(f"Séparé document en {len(all_splits)} sous-documents.")
```

```python
import os
import logging
from pinecone import Pinecone
from langchain.vectorstores import Pinecone as PineconeLangChain
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

# 🔹 Configuration des logs
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# 🔹 Clés API (directement dans le code)
PINECONE_API_KEY = "pcsk_53U2S4_H35UkcZBx97Wr6JgdNpjhb6yKHEhr8XHDUFTwUtyQbjHk5AYxmSEPLBf9VHBNzC"
GOOGLE_API_KEY = "AIzaSyAzPkQdUlVtp3iqre6lxMACoPUQNFKlYJg"

# 🔹 Set the Pinecone API key as an environment variable
os.environ["PINECONE_API_KEY"] = PINECONE_API_KEY

# 🔹 Initialisation de Pinecone
pc = Pinecone(api_key=PINECONE_API_KEY)

# 🔹 Création de l'index Pinecone (si nécessaire)
INDEX_NAME = "pdf-documents-index-v2"

if INDEX_NAME not in pc.list_indexes().names():
    pc.create_index(name=INDEX_NAME, dimension=1536, metric="cosine")

# 🔹 Initialisation du modèle d'embeddings
embedding_model = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-large")

# 🔹 Chargement de Pinecone via LangChain (sans le paramètre `pinecone_client`)
vectorstore = PineconeLangChain.from_existing_index(
    index_name=INDEX_NAME,
    embedding=embedding_model
)

# 🔹 Définition du modèle LLM (Google Gemini via LangChain)
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-pro",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.5,
    max_tokens=None,
    max_retries=2
)

# 🔹 Prompt optimisé avec un template LangChain
prompt_template = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    Utilise le contexte suivant pour répondre à la question de l'utilisateur de manière claire et concise.

    Contexte :
    {context}

    Question :
    {question}

    Réponse :
    """
)

# 🔹 Chaîne de recherche et de génération
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)

# 🔹 Fonction pour répondre aux questions
def answer_question(question: str) -> str:
    try:
        logging.info("🔍 Recherche et génération de réponse...")
        result = qa_chain({"query": question})
        response = result["result"]
        sources = result.get("source_documents", [])

        if not response.strip():
            logging.warning("⚠️ Aucune réponse générée.")
            return "Désolé, aucune réponse pertinente n'a été trouvée."

        # 🔹 Afficher les sources utilisées
        if sources:
            source_texts = "\n".join([doc.page_content[:200] for doc in sources])  # Affiche les 200 premiers caractères
            logging.info(f"📜 Documents sources récupérés :\n{source_texts}")
        else:
            logging.info("⚠️ Aucun document source trouvé.")

        return response.strip()
    except Exception as e:
        logging.error(f"❌ Erreur : {e}")
        return "Une erreur s'est produite. Veuillez réessayer."

# 🔹 Test interactif
if __name__ == "__main__":
    try:
        user_question = input("Entrez votre question : ")
        response = answer_question(user_question)
        print("\n🤖 Réponse de Gemini :")
        print(response)
    except KeyboardInterrupt:
        print("\n⏹️ Programme arrêté par l'utilisateur.")
    except Exception as e:
        logging.error(f"❌ Erreur inattendue : {e}")
```
